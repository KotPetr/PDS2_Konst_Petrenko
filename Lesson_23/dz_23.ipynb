### Task 1
1. Прилад містить 7 елементів, з яких три зношені. При включенні приладу випадковим чином включаються три елементи. 
Знайти ймовірність того, що при цьому виявляться включеними два незношені елементи.

P(A) - ймовірність події  
Р(А') - ймовірність випадку, коли із включених елементів 2 незношені, а один - зношений.   
Таких варіантів існує три: 110, 101, 011.  
Для всіх трьох варіантів ймовірність однакова, тобто  
Р(А') = (4/7 * 3/6 * 3/5) = (3/7 * 4/6 * 3/5) = 0.17 , таким чином:    
Р(А) = Р(А') * 3 
P(A) = 0.17 * 3   
P(A) = **0,51**  

### Task 2

2. Для сигналізації про аварію встановлено два незалежно працюючих сигналізатори. 
Ймовірність того, що при аварії спрацює перший сигналізатор дорівнює 0,95, ця ймовірність для другого сигналізатора 
дорівнює 0,9. Знайти ймовірність того, що при аварії спрацює:   
а) хоча б один сигналізатор;  
б) тільки один сигналізатор.

+ а) Ймовірність того, що не спрацює жоден сигналізатор дорвнює:  
Р(А') = (1 - 0.9) * (1 - 0.95) = 0.005  
Таким чином, ймовірність того, що спрацює хочаб один сигналізатор:  
Р(А) = 1 - 0.005 = **0.995**
+ б)  Ймовірність того що спарцює тільки один сигналізатор є об'єднання перетинів двох подій: спрацьовує перший сигналізатор і не спрацьовує другий і навпаки, таким чином:  
Р(А) = Р(А) * Р(В') + Р(В) * Р(А')  
Р(А) = 0.9 * 0.05 + 0.95 * 0.1 = **0.14**
    

### Task 3

3. Два верстати-автомати виготовляють однакові вироби, які попадають на загальний конвеєр. Продуктивність першого автомата 
в два раза більш ніж продуктивність другого. Перший верстат-автомат виготовляє 60% виробів вищого гатунку, а другий - 84 %.   
а) Знайти ймовірність того, що навмання взятий з конвеєра виріб виш ґатунку.  
б) Навмання взятий з конвеєра виріб має вищий ґатунок. Знайти ймовірність того, що він виготовлений на першому верстаті.

+ а) Ймовірність того, що навмання взятий з конвеєра виріб вироблено на першому автоматі 2/3, а на другому 1/3  
Ймовірність того, що виріб буде вищого гатунку:  
P(A) = (P(A1) * 2/3 + P(A2) * 1/3)  
Р(А) = (0.6) * 2/3 + 0.84 * 1/3  
Р(А) = **0.68**


+ б) В умовну одиницю часу перший автомат виготовляє 60 + 60 виробів вищого гатунку, за цей же час другий автомат виробляє 84 виробів вищого гатунку. Тобто з 204 виробів 120 виготовлені на першому верстаті.  
P(A1) = 120/204  
Р(А1) = **0.59**
