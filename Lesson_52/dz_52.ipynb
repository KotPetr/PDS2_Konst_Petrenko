## Завдання:
Використовуючи клас GaussianNB з біблотеки Scikit-Learn створіть
модель класифікації для навчання з використанням датасету Iris (sklearn.datasets.load_iris). Після цього, за допомогою методів з бібліотеки ml_metrics, обрахуйте метрики для отриманої моделі.


```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
```

```python
data = load_iris()
X, y, column_names = data['data'], data['target'], data['feature_names']
X = pd.DataFrame(X, columns = column_names)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=5, stratify=y)
```

```python
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_val)
df = pd.DataFrame({'y_test' : y_val, 'y_pred': y_pred})
df['is_equal'] = df['y_test'] == df['y_pred']
df
```

```python
cnf_matrix = metrics.confusion_matrix (y_val, y_pred)
print('Confusion matrix: ')
pd.DataFrame(cnf_matrix)
```

```python
print(f'Accuracy: {metrics.accuracy_score(y_val, y_pred):.4f}')
```

### Висновок:  
При додаванні параметра **stratify**  у функцію розбивки **train_test_split** якість моделі дещо покращується.   
Також на якість даної моделі впливає параметр **random_state**, що взагалі не коректно. На мій погляд, це пов'язано з невиликим об'ємом тестової вибірки.
