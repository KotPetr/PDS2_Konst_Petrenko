### Task 2
У рибальскому містечку 99,99% чоовіків хоча б раз були на рибалці. Проводять соціологічні дослідження серед 10000 навмання вибраних чоловіків. Визначити дисперсію D(X) і середнє квадратичне відхилення S(X) випадквої величини X - числа чоловіків, які жодного разу не були на рибалці.


В даному випадку ми має маємо справу з розподілом Пуассона, оскільки ймовірність події, яка нас цікавить, наближається до 0, а число випробувань досить велике.      
$р(Х) = 1 - (99.99 / 100) = 0.0001$;  
$n = 10000$;  
$np = 1000 * 0.0001 = 1 \leq 10$  
Тому показник дисперсії можна розрахувати за формулою: $D(X) = np$  
$D(X) = 10000 * 0.0001 = 1$  
Звідси среднє квадратичне відхилення буде дорівнювати $S(X) = \sqrt{D(X)}$   
$S(X) = \sqrt{1} = 1$ 

**Відповідь:** Дисперсія D(X) = **1**, середнє квадратичне відхилення S(X) = **1**.
