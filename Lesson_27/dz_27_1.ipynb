### Task 1
Мікропроцесор має 10000 транзисторів, які працюють незалежно один від одного. Імовірність того, що транзистор вийде з ладу під час роботи, є величиною малоймовірною і становить 0,0007. Визначити математичне очікування M(X) і середнє квадратичне відхилення S(X) випадкової величини X - числа транзисторів, що вийдуть із ладу під час роботи мікропроцесора.



В даному випадку ми має маємо справу з розподілом Пуассона, оскільки ймовірність події, яка нас цікавить, наближається до 0, а число випробувань досить велике.    
$р(Х) = 0.0007$;  
$n = 10000$;  
$np = 10000 * 0.0007 = 7 \leq 10$  
Тому показник математичного очікування можна розрахувати за формулою: $M(X) = np$  
$M(X) = 10000 * 0.0007 = 7$  
У випадку Пуассоновського розподілу математичне очікування дорівнює дисперсії $M(X) = D(X) = np$  
Тобто D(X) = 7  
Звідси среднє квадратичне відхилення буде дорівнювати $S(X) = \sqrt{D(X)}$   
$S(X) = \sqrt{7} = 2.65$ 

**Відповідь:** Математичне очікування М(X) = **7**, середнє квадратичне відхилення S(X) = **2.65**.
