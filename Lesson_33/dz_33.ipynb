```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

#### 1. Створення датафрейму з .xlsx файлу 

```python
df = pd.read_excel('cars.xlsx')
df
```

#### 2. Візуалізація залежностей

```python
fig = plt.figure(figsize=(22, 6), dpi=80)
plt.style.use('bmh')

ax = fig.add_subplot(131)
price = df.sort_values(by='price_usd').price_usd
fuel_cons = df.sort_values(by='price_usd').fuel_cons
plt.title('Залежність витрати пального від ціни авто')
plt.xlabel('Ціна авто, usd')
plt.ylabel('Витрата пального, л/100 км')
plt.plot(price, fuel_cons, color= 'magenta', linewidth = 2)


ax = fig.add_subplot(132)
max_speed = df.sort_values(by='max_speed').max_speed
fuel_cons = df.sort_values(by='max_speed').fuel_cons
plt.title('Залежність витрати пального від макс. швидкості')
plt.xlabel('Максимальна швидкість, км/год')
plt.ylabel('Витрата пального, л/100 км')
plt.scatter(max_speed, fuel_cons, color='blue', marker = 'o')

ax = fig.add_subplot(133)
max_speed = df.sort_values(by='max_speed').max_speed
load_cap = df.sort_values(by='max_speed').load_cap
plt.title('Залежність вантажопідйомності від макс. швидкості')
plt.xlabel('Максимальна швидкість, км/год')
plt.ylabel('Вантажопідйомність, кг')
plt.scatter(max_speed, load_cap, color='blue', marker = 'd' )
```

#### 3. Гісторгама групування моделей авто за максимальною швидкістю

```python
max_speed = df.max_speed
plt.title('Кількість авто по групам за максимальною швидкістю.', fontsize = 12)
plt.xlabel('Максимальна швидкість, км/год')
plt.ylabel('Кількість моделей авто, шт.')

plt.hist(max_speed, 5, color='orange', width = 17)
```

#### 4. Коробковий графік розподілення авто за максимальною швидкістю

```python
sb.boxplot(data=df, y='max_speed', color= 'cyan')
sb.set_style('darkgrid')
plt.ylabel('Максимальна швидкість, км/год')
plt.title('Розподілення авто за макс.швидкістю.')
```

#### 5. Графіки попарних залежностей даних датафрейму.

```python
sb.pairplot(df)
```
