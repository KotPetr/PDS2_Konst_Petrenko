```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_excel('cars.xlsx')
df
```

```python
fig = plt.figure(figsize=(20, 6), dpi=80)

ax = fig.add_subplot(131)
price = df.sort_values(by='price_usd').price_usd
fuel_cons = df.sort_values(by='price_usd').fuel_cons
plt.title('Залежність витрати пального від ціни авто')
plt.xlabel('Ціна авто, usd')
plt.ylabel('Витрата пального, л/100 км')
plt.plot(price, fuel_cons, color= 'magenta', linewidth = 2)


ax = fig.add_subplot(132)
max_speed = df.sort_values(by='max_speed').max_speed
fuel_cons = df.sort_values(by='max_speed').fuel_cons
plt.title('Залежність витрати пального від макс. швидкості')
plt.xlabel('Максимальна швидкість, км/год')
plt.ylabel('Витрата пального, л/100 км')
plt.scatter(max_speed, fuel_cons, color='blue', marker = 'o')

ax = fig.add_subplot(133)
max_speed = df.sort_values(by='max_speed').max_speed
load_cap = df.sort_values(by='max_speed').load_cap
plt.title('Залежність вантажопідйомності від макс. швидкості')
plt.xlabel('Максимальна швидкість, км/год')
plt.ylabel('Вантажопідйомність, кг')
plt.scatter(max_speed, load_cap, color='purple', marker = 'd' )
```
