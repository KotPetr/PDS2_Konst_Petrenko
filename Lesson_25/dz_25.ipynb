### Task 1

Ймовірність того, що при перевірці деталь виявиться стандартною, дорівнює 0,6. Перевірці підлягають 3 деталі. Побудувати закон розподілу системи двох дискретних випадкових величин X - появи числа бракованих деталей і появи числа стандартних деталей.

Хай Х - подія появи стандарних деталей, яка трапляється з ймовірністю p = 0,6,  
а Y - подія появи бракованих деталей, трапляється з ймовірністю q = 1 - 0,6 = 0,4.  
Ймовірність появи із n деталей k стандартних буде рахуватись за формулою:    

$$ x_iy_j = C_n^k p^xq^y= \frac{n!}{k!(n-k)!} p^xq^y$$  

Тоді закон розподілу можна представити в вигляді наступної таблиці:

Y \X|  0  |  1  |  2  |  3  | P(Y)|
:--:|:---:|:---:|:---:|:---:|:---:|
 0  |  0  |  0  |  0  |0.064|0.064|                                   
 1  |  0  |  0  |0.288|  0  |0.288|
 2  |  0  |0.432|  0  |  0  |0.432|
 3  |0.216|  0  |  0  |  0  |0.216|
P(X)|0.216|0.432|0.288|0.064|  1  |

Розрахунки статистичних коефіціентів:

```python
import numpy as np
```

```python
X = np.array([0, 1, 2, 3])
Y = np.array([0, 1, 2, 3,])
PXY= np.array([[0, 0, 0, 0.064],
               [0, 0, 0.288, 0],
               [0, 0.432, 0, 0],
               [0.216, 0, 0, 0]])

PX = [round(sum(PXY[i]), 3) for i in range(len(PXY))]
PY = [round(sum(PXY[:, i]), 3) for i in range(len(PXY[0]))]   

MX = sum(X*PX)              # Математичне очікування
DX = sum(X**2*PX) - MX**2   # Дисперсія
sX = DX**0.5                # Стандартне відхилення
print(f'Математичне очікування появи стандартної деталі Х дорівнює: {round(MX,2)}. Дисперсія: {round(DX,4)}. '+ 
      f'Стандарте відхилення: {round(sX, 4)}')

MY = sum(Y*PY)              # Математичне очікування
DY = sum(Y**2*PY) - MY**2   # Дисперсія
sY = DY**0.5                # Стандартне відхилення
print(f'Математичне очікування появи бракованої деталі Y дорівнює: {round(MY,2)}. Дисперсія: {round(DY,4)}. '+ 
      f'Стандарте відхилення: {round(sY, 4)}')

M_XY = 0  
for i in range(len(X)):
    M_XY += X[i]*(sum(PXY[i]*Y))  # M(XY)

print(f'Математичне очікування добутку XY дорівнює {M_XY}')

covXY = M_XY - MX * MY            # коефіціент коваріації
print(f'Коефіціент коваріації дорівнює {round(covXY, 4)}')

rXY = covXY/(sX*sY)               # коефіціент кореляції
print(f'Коефіціент кореляції дорівнює {round(rXY)}')
```

Коефіціент кореляції дорівнює -1 - це є повна відємна кореляція, так як поява стандартної деталі абсолютно виключає появу бракованої деталі і навпаки.

Розглянемо дві різні умовні акції X і У (табл. на наступному слайді). Для кожної з них за допомогою
статистичних досліджень були розраховані відносні частоти очікуваних прибутків залежно від п'яти 
станів економічного середовища: значне піднесення (ЗП), незначне піднесення (НЗП), стагнація (C), 
незначний спад (НЗС), значний спад (ЗС). Ці дані занесені відповідно в останній рядок і останній 
стовпчик таблиці 1. Залежно від економічного середовища прогнозовані норми очікуваних прибутків у 
відсотках задані відповідно у першому рядку та першому стовпчику таблиці.   
У клітинках на перетині X і У проставлені ймовірності появи HOPM відсотків значень (x, y).


![](table1.png 'Venn diagram')

Необхідно:
1. обчислити очікувану норму прибутку по кожній з акцій;
2. оцінити ступінь ризику обох акцій за допомогою середньоквадратичного відхилення;
3. проаналізувати наявність зв'язку між акціями Х та V за допомогою коефіцієнтів коваріації та кореляції.



```python
X = np.array([10, 5, 2, 1, -5])
Y = np.array([20, 10, 2, -2, -8])
PXY= np.array([[0.01, 0.03, 0.02, 0.03, 0.01], 
               [0.03, 0.09, 0.06, 0.09, 0.03],
               [0.02, 0.06, 0.04, 0.06, 0.02],
               [0.03, 0.09, 0.06, 0.09, 0.03],
               [0.01, 0.03, 0.02, 0.03, 0.01]])

PX = [round(sum(PXY[:, i]), 2) for i in range(len(PXY[0]))]   # 0.1 ,0.3, 0.2, 0.3, 0.1
PY = [round(sum(PXY[i]), 2) for i in range(len(PXY))]         # 0.1 ,0.3, 0.2, 0.3, 0.1

MX = sum(X*PX)             # Математичне очікування
DX = sum(X**2*PX) - MX**2  # Дисперсія
sX = DX**0.5                # Стандартне відхилення
print(f'Математичне очікування прибутку від акції Х дорівнює: {round(MX,2)}. Дисперсія: {round(DX,4)}. '+ 
      f'Стандарте відхилення: {round(sX, 4)}')

MY = sum(Y*PY)              # Математичне очікування
DY = sum(Y**2*PY) - MY**2   # Дисперсія
sY = DY**0.5                # Стандартне відхилення
print(f'Математичне очікування прибутку від акції Y дорівнює: {round(MY,2)}. Дисперсія: {round(DY,4)}. '+ 
      f'Стандарте відхилення: {round(sY, 4)}')

M_XY = 0  
for i in range(len(X)):
    M_XY += X[i]*(sum(PXY[i]*Y))  # M(XY)

print(f'Математичне очікування добутку XY дорівнює {M_XY}')

covXY = M_XY - MX * MY            # коефіціент коваріації
print(f'Коефіціент коваріації дорівнює {covXY}')

rXY = covXY/(sX*sY)               # коефіціент кореляції
print(f'Коефіціент кореляції дорівнює {rXY}')

```

На основі проведених розрахунків можна зробити наступні висновки:
+ Очікуваний прибуток від акції Y більший ніж від акції Х 
+ В той же час акція Y є більш ризикова ніж акція Х
+ Поведінка акцій Х та Y не пов'язана між собою.
