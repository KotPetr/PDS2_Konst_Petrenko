### Task 2

При вивченні випадкової величини Х у результаті 40 незалежних спостережень дістали вибірку: 10, 13, 10, 9, 9, 12, 12, 6, 7, 9, 8, 9, 11, 9, 14, 13, 9, 8, 8, 7, 10, 10, 11, 11, 11, 12, 8, 7,9, 10, 14, 13, 8, 8, 9, 10, 11, 11, 12, 12. Побудувати дискретний статистичний розподіл для цієї вибірки, а також полігон частот і F*(x). Знайти моду та медіану.

```python
import numpy as np
from prettytable import PrettyTable 
import matplotlib.pyplot as plt
```

```python
x = [10, 13, 10, 9, 9, 12, 12, 6, 7, 9, 8, 9, 11, 9, 14, 13, 9, 8, 8, 7, 
    10, 10, 11, 11, 11, 12, 8, 7, 9, 10, 14, 13, 8, 8, 9, 10, 11, 11, 12, 12]

X = np.array(x)
n = len(X)

X_set= list(set(x))
frec = [x.count(i) for i in X_set]
rel_frec = [i/n for i in frec]

moda = X_set[frec.index(max(frec))]
median = np.median(X)

print('Об\'єм вибірки: ', n )

print("Дискретний статистичний розподіл для Х має вигляд: \n")
head = X_set
row1 = frec
row2 = rel_frec
head.insert(0, 'Варіанта')
row1.insert(0, 'Частота')
row2.insert(0, "Відносна частота")

table = PrettyTable(head)
table.add_row(row1)
table.add_row(row2)

print(table)

print()
print('Медиана для Х: ', median)
print("Мода для Х: ", moda)
```

#### Полігон частот:

```python
plt.plot(X_set, frec, marker='D')
plt.grid()
plt.xlabel('Варіанта')
plt.ylabel('Частота')
plt.show()
```

#### Полігон відносних частот:

```python
plt.plot(X_set, rel_frec, marker='D')
plt.grid()
plt.xlabel('Варіанта')
plt.ylabel('Відносна частота')
plt.show()
```
